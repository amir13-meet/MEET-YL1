```python
import csv
f = open('cig.csv', 'rb')
r = csv.reader(f)
mf_tar = []
mf_nicotine = []

for i in r:
    x = i[0].split()
    x = x[1:]
    x = map(float, x)
    mf_tar.append(x[0])
    mf_nicotine.append(x[1])
  
f.close()

from pylab import *

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Some Mawfuckin Shit About Cigarettes')
ax.set_xlabel("Mawfuckin Tar")
ax.set_ylabel('Mawfuckin Nicotine')
scatter(mf_tar, mf_nicotine)

savefig("mf_cig_scatter.png")

```
