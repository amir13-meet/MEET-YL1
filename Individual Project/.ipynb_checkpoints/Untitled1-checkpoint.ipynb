This is code and a graph about environmental stuff. It came from: 
"http://lib.stat.cmu.edu/datasets/visualizing.data.zip"

```python
import csv
y = open('environmental.csv', 'rb')
x = csv.reader(y)
rownum = 0
temperature = []
wind = []

for i in x:
    if (rownum > 0):
        z = map(float, i)
        temperature.append(z[2])
        wind.append(z[3])
    rownum += 1

y.close()

from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The Relationship between Wind and Temperature')
ax.set_xlabel('Temperature')
ax.set_ylabel('Wind')
bar(temperature, wind)
savefig("environmental_bar")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The Relationship between Wind and Temperature')
ax.set_xlabel('Temperature')
ax.set_ylabel('Wind')
scatter(temperature, wind)
savefig("environmental_scatter")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The Relationship between Wind and Temperature')
ax.set_xlabel('Temperature')
ax.set_ylabel('Wind')
plot(temperature, wind)
savefig("environmental_plot")
```

This graph shows the relationship between Wind and Temperature
